```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': ['a', 'b', 'c']})
df2 = pd.DataFrame({'A': [2, 3, 4], 'C': [10, 20, 30]})

# Inner join
inner_join_df = pd.merge(df1, df2, how='inner', on='A')

# Left join
left_join_df = pd.merge(df1, df2, how='left', on='A')

# Right join
right_join_df = pd.merge(df1, df2, how='right', on='A')

# Outer join
outer_join_df = pd.merge(df1, df2, how='outer', on='A')

# Cross join
cross_join_df = pd.merge(df1, df2, how='cross')

# Print the resulting DataFrames
print("Inner Join:\n", inner_join_df)
print("Left Join:\n", left_join_df)
print("Right Join:\n", right_join_df)
print("Outer Join:\n", outer_join_df)
print("Cross Join:\n", cross_join_df)
```
