```python
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the data
plt.plot(x, y, label='Line')
plt.axhline(y=3, color='red', linestyle='--', label='Vertical Line')

# Add legend with adjusted vertical spacing
plt.legend(labelspacing=0.5)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(2, 2)

# Plot data on each subplot
axs[0, 0].plot([1, 2, 3], [4, 5, 6], label='Line 1')
axs[0, 1].plot([1, 2, 3], [2, 4, 6], label='Line 2')
axs[1, 0].plot([1, 2, 3], [3, 6, 9], label='Line 3')
axs[1, 1].plot([1, 2, 3], [1, 3, 5], label='Line 4')

# Create a single legend
fig.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a Pandas DataFrame
data = pd.DataFrame({'group': ['A', 'A', 'B', 'B', 'C', 'C'], 'values': [1, 2, 3, 4, 5, 6]})

# Plot a box plot using Pandas
data.boxplot(column='values', by='group')

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Create a Pandas DataFrame
data = pd.DataFrame({'group': ['A', 'A', 'B', 'B', 'C', 'C'], 'values': [1, 2, 3, 4, 5, 6]})

# Plot a box plot using Seaborn
sns.boxplot(data=data, x='group', y='values')

# Show the plot
plt.show()
```
